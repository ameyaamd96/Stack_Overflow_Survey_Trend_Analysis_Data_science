```python
possible_vals = ['Java','JavaScript','CSS','PHP','Python','Ruby','SQL','C#','C++','C','Perl', 'HTML5','Node.js','Objective-C', 'AngularJS', 'Swift', 'Scala', 'Rust']

def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]

    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up
    '''
    new_df = defaultdict(int)
    #loop through list of ed types
    for val in look_for:
        #loop through rows
        for idx in range(df.shape[0]):
            #if the ed type is in the row add 1
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values('count', ascending=False, inplace=True)
    return new_df

def clean_and_plot(df, title='Language_dis_yearly'):
    '''
    INPUT 
        df - a dataframe holding the Language used column
        title - string the title of your plot
        axis - axis object
        plot - bool providing whether or not you want a plot back
        
    OUTPUT
        study_df - a dataframe with the count of how many individuals
        Displays a plot of pretty things related to the CousinEducation column.
    '''
    study = df["Language_used"][1:].value_counts().reset_index()
    study.rename(columns={'index': 'Language', "Language_used": 'count'}, inplace=True)
    study_df = total_count(study, 'Language', 'count', possible_vals)

    study_df.set_index('Language', inplace=True)
    props_study_df = study_df/study_df.sum()
    return props_study_df



df_2011["Language_used"] = df_2011[['Which languages are you proficient in?', 'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 36', 'Unnamed: 37', 'Unnamed: 38', 'Unnamed: 39', 'Unnamed: 40', 'Unnamed: 41', 'Unnamed: 42']].astype(str).agg(','.join, axis=1)    
props_df_2011 = clean_and_plot(df_2011)
props_df_2011 = props_df_2011.rename(columns={'count': '2011'})

df_2012["Language_used"] = df_2012[['Which languages are you proficient in?', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 36']].astype(str).agg(','.join, axis=1)
props_df_2012 = clean_and_plot(df_2012)
props_df_2012 = props_df_2012.rename(columns={'count': '2012'})

df_2013["Language_used"] = df_2013[['Which of the following languages or technologies have you used significantly in the past year?', 'Unnamed: 57', 'Unnamed: 58', 'Unnamed: 59', 'Unnamed: 60', 'Unnamed: 61', 'Unnamed: 62', 'Unnamed: 63', 'Unnamed: 64', 'Unnamed: 65', 'Unnamed: 66', 'Unnamed: 67', 'Unnamed: 68', 'Unnamed: 69']].astype(str).agg(','.join, axis=1)
props_df_2013 = clean_and_plot(df_2013)
props_df_2013 = props_df_2013.rename(columns={'count': '2013'})

df_2014["Language_used"] = df_2014[['Which of the following languages or technologies have you used significantly in the past year?', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50', 'Unnamed: 51', 'Unnamed: 52', 'Unnamed: 53']].astype(str).agg(','.join, axis=1)
props_df_2014 = clean_and_plot(df_2014)
props_df_2014 = props_df_2014.rename(columns={'count': '2014'})

df_2015.columns=df_2015.iloc[0]
df_2015["Language_used"] = df_2015[['Current Lang & Tech: AngularJS', 'Current Lang & Tech: C', 'Current Lang & Tech: C++', 'Current Lang & Tech: C++11', 'Current Lang & Tech: C#', 'Current Lang & Tech: Go', 'Current Lang & Tech: Java', 'Current Lang & Tech: JavaScript', 'Current Lang & Tech: Node.js', 'Current Lang & Tech: Objective-C', 'Current Lang & Tech: Perl', 'Current Lang & Tech: Python', 'Current Lang & Tech: PHP', 'Current Lang & Tech: R', 'Current Lang & Tech: Ruby', 'Current Lang & Tech: Rust', 'Current Lang & Tech: SQL', 'Current Lang & Tech: Rust', 'Current Lang & Tech: Scala']].astype(str).agg(','.join, axis=1)
props_df_2015 = clean_and_plot(df_2015)
props_df_2015 = props_df_2015.rename(columns={'count': '2015'})

df_2016["Language_used"] = df_2016['tech_do']
props_df_2016 = clean_and_plot(df_2016)
props_df_2016 = props_df_2016.rename(columns={'count': '2016'})

df_2017["Language_used"] = df_2017['HaveWorkedLanguage']
props_df_2017 = clean_and_plot(df_2017)
props_df_2017 = props_df_2017.rename(columns={'count': '2017'})

df_2018["Language_used"] = df_2018['LanguageWorkedWith']
props_df_2018 = clean_and_plot(df_2018)
props_df_2018 = props_df_2018.rename(columns={'count': '2018'})

df_2019["Language_used"] = df_2019['LanguageWorkedWith']
props_df_2019 = clean_and_plot(df_2019)
props_df_2019 = props_df_2019.rename(columns={'count': '2019'})

df_2020["Language_used"] = df_2020['LanguageWorkedWith']
props_df_2020 = clean_and_plot(df_2020)
props_df_2020 = props_df_2020.rename(columns={'count': '2020'})

df_2021["Language_used"] = df_2021['LanguageHaveWorkedWith']
props_df_2021 = clean_and_plot(df_2021)
props_df_2021 = props_df_2021.rename(columns={'count': '2021'})

df_2022["Language_used"] = df_2022['LanguageHaveWorkedWith']
props_df_2022 = clean_and_plot(df_2022)
props_df_2022 = props_df_2022.rename(columns={'count': '2022'})
```

```python
def monthly_yearly_clean(df):
  for i in range(len(df['CompTotal'])):
    if df['CompFreq'][i] == 'Monthly':
      if df['Yearly_Salary'][i] > 50000:
        df['Yearly_Salary'][i] = np.nan
      else: 
        df['Yearly_Salary'][i] *= 12
    elif df['CompFreq'][i] == 'Yearly':
        if df['Yearly_Salary'][i] > 600000:
          df['Yearly_Salary'][i] = np.nan
        else:
          df['Yearly_Salary'][i] = df['Yearly_Salary'][i]
    else:
      df['Yearly_Salary'][i] = np.nan
  return df

df_2021 = monthly_yearly_clean(df_2021)
df_2022 = monthly_yearly_clean(df_2022)
```

```python
def salary_midpoint_cal(df):
  for i in range(len(df['Including bonus, what is your annual compensation in USD?'])):
    if df['Including bonus, what is your annual compensation in USD?'][i] == '<$20,000':
      df['Yearly_Salary'][i] = 10000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '$20,000 - $40,000':
      df['Yearly_Salary'][i] = 30000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '$40,000 - $60,000':
      df['Yearly_Salary'][i] = 50000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '$60,000 - $80,000':
      df['Yearly_Salary'][i] = 70000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '$80,000 - $100,000':
      df['Yearly_Salary'][i] = 90000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '$100,000 - $120,000':
      df['Yearly_Salary'][i] = 110000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '$120,000 - $140,000':
      df['Yearly_Salary'][i] = 130000
    elif df['Including bonus, what is your annual compensation in USD?'][i] == '>$140,000':
      df['Yearly_Salary'][i] = 150000  
    else:
      df['Yearly_Salary'][i] = np.nan
  return df

df_2011 = salary_midpoint_cal(df_2011)
df_2012 = salary_midpoint_cal(df_2012)
df_2013 = salary_midpoint_cal(df_2013)
df_2014 = salary_midpoint_cal(df_2014)
```
